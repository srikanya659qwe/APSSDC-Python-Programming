```python
# Function to print the nearest to zero in array (if multiple nos are there print greater value)
li=[-1,-2,2,3]
li.sort()
pl=[]
for i in li:
    pl.append(abs(i))
pl.sort()
if pl[0] in li:
    print(pl[0])
else:
    print(-pl[0])
    
```

### Problem3
- You are given three numbers, a,b and c.
- Write a program to find the largest number
- which is less than or equal to  c and leaves
- remainder b when divided by a


```python
def largestnumber(a,b,c):
    for i in range(c,a,-1):  #c,c-1,...a
        if i % a == b:
            return i
    return -1
        
largestnumber(3,2,9)
```

```python
#Prime numbers
def prime(n):
    count=0
    for i in range(1,n+1):
        if n%i==0:
            count=count+1
    if count==2:
        return True
    else:
        return False
        
def isprime(a):
    for i in range(1,a):
        if prime(i):
            print(i,end=" ")


n=int(input())
isprime(n)
    
```

```python
### Two Strings

n=int(input())
def twostrings(f,s):
    f1=1
    if len(f)!=len(s):
        return 'NO'
    else:
        for i in range(len(f)):
            if f.count(f[i])!=s.count(f[i]):
                return 'NO'
        if f1==1:
                return 'YES'
for i in range(n):
    st=input().split()
    f=st[0]
    s=st[1]
    print(twostrings(f,s))
```

```python
### Duration
def duration(sh,eh,sm,em):
    a=(sh*60)+sm
    b=(eh*60)+em
    c=b-a
    d=c//60
    e=c%60
    print(d,end=" ")
    print(e,end=" ")
duration(1,2,44,14)
duration(2,8,42,23)
    


```
